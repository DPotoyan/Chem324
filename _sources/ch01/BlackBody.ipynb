## DEMO: Black Body radiation

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def planck(wavelength, T):
    """
    Planck's Law to calculate the spectral radiance of black body radiation at temperature T.
    
    Parameters:
    - wavelength (float): Wavelength in meters.
    - T (float): Absolute temperature in Kelvin.
    
    Returns:
    - (float): Spectral radiance in W/(m^2*sr*m).
    """
    
    from scipy.constants import h, c, k
    
    return (2.0*h*c**2) / (wavelength**5 * (np.exp((h*c)/(wavelength*k*T)) - 1))
```

```python
 # Wavelength range 
wavelengths = np.linspace(1e-9, 5e-6, 1000) 
  
for T in [3000, 4000, 5000]:
    
    plt.plot(1e9*wavelengths, planck(wavelengths, T), label=f'T={T}')
  
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity ($W/m^2/nm$)')
plt.title('Black Body Radiation Curve')
plt.legend()
plt.ylim(1,    1.5e13)
plt.grid()
plt.show()
```

### Interactive plot of black body radiation

```python
@interact(T=(1000,10000))
def interactive_planck(T):
    """
    Planck's Law to calculate the spectral radiance of black body radiation at temperature T.
    
    Parameters:
    - T (float): Absolute temperature in Kelvin.
    
    Returns:
    - (float): Spectral radiance in W/(m^2*sr*m).
    """
    
    from scipy.constants import h, c, k
    
    wavelengths = np.linspace(1e-9, 5e-6, 1000) 
    
    intensity= (2.0*h*c**2) / (wavelengths**5 * (np.exp((h*c)/(wavelengths*k*T)) - 1))
    
    plt.plot(1e9*wavelengths, intensity, label=f'T={T}')
    plt.legend()
    plt.show()
```
